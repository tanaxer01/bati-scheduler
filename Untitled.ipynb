```python
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re

schedule_names = ['FCFS', 'EASY', 'DQN']

pd.DataFrame.append = pd.DataFrame._append

BASE_DIR = Path("./expe-out")
```

```python
sim_files  = BASE_DIR.glob("sim-*.out")
jobs_files = BASE_DIR.glob("jobs-*.out")
schedule_files = BASE_DIR.glob("schedule-*.out")
```

```python
metrics = pd.DataFrame()
for file in BASE_DIR.glob("schedule-*.out"):
    algorithm = re.search(r'schedule-(.*?)\.', str(file)).group(1)

    curr = pd.read_csv(str(file))
    curr["schedule"] = algorithm
    metrics = metrics.append(curr)

metrics.set_index('schedule', inplace=True)
metrics
```

```python
from evalys.jobset import JobSet
jobsets = {}

for file in BASE_DIR.glob("jobs-*.out"):
    algorithm = re.search(r'jobs-(.*?)\.', str(file)).group(1)
    jobsets[algorithm] = JobSet.from_csv(str(file))
```

```python
def create_metric_df(metric):
    return pd.concat([jobsets[schedule].df[metric].rename(schedule) for schedule in schedule_names], axis=1)
```

```python
dqn  = JobSet.from_csv("expe-out/jobs-DQN.out")
fcfs = JobSet.from_csv("expe-out/jobs-FCFS.out")
easy = JobSet.from_csv("expe-out/jobs-EASY.out")

dqn.plot(with_details=True)
plt.show()
#from evalys.visu.series import UtilizationSeriesVisualization
#UtilizationSeriesVisualization(dqn, title="DQN")
```

```python
waiting_times = create_metric_df('waiting_time')
waiting_times.describe()
```
